```python
import numpy as np

n_elem = 3092691

# Find the closest factor of n_elem
factors = np.array([i for i in range(1, int(np.sqrt(n_elem)) + 1) if n_elem % i == 0])
n_lat = factors[np.argmin(np.abs(factors - int(np.sqrt(n_elem))))]
n_lon = int(n_elem / n_lat)

# Create a one-dimensional numpy array
data_1d = np.arange(n_elem)

# Reshape the one-dimensional array into a two-dimensional array
data_2d = np.reshape(data_1d, (n_lat, n_lon))
```
